```python
import pandas as pd
import requests
import io
import seaborn
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/jehilm/nba-pitch/master/2010/players.csv?token=AMDNULIFI3E2TIKRTRRSPRLANMZSW"
download = requests.get(url).content
data = pd.read_csv(io.StringIO(download.decode('utf-8')))
```
